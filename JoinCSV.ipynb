```python
import pandas as pd; import numpy as np
clean1 = pd.DataFrame.from_csv('clean1.csv')
relative = pd.Series.from_csv('relative diversity.csv')
```

```python
Relative = []

for i in clean1.index:
    try:
        Relative.append(relative[i]) 
    except:
        Relative.append(np.NaN) 
clean1['R'] = Relative
clean1.dropna()
```

```python
clean1.to_csv('clean1R.csv')
```
