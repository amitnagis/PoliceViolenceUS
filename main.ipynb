# Understanding police violence in United States

## Introduction

We present here an exploratory, on-going project to understand and model the pattern of police violence across United States. By police violence, we do not necessarily mean police brutality is associated which is associated with indiscriminate and disproportionate use of force by the police on common citizenry. In other words, instances of police brutality would be a subset of instances of police violence. Here we consider a subset of instances of police violence, namely those instances when police officers resorted to use of deadly force. This choice is motivated mainly by the available extensive, good-quality data. Our goal is to study the instances of police violence in terms of its geospatial distribution and it's correlation (or lack thereof) with other intuitive features such as local crime rate in the jurisdiction. Moreover, we would like to understand if police diversity (as an independent parameter and in relation to local population diversity) is correlated with use of deadly force by the police.

## Quantifying Diversity

The databases mentioned above have detailed information about percentage population of various racial groups. In our first attempt we coarse-grain our information and use one measure for diversity insensitive to the racial make-up of the same. To this end we introduce a quantity, information entropy ($D$) to quantify diversity in general population and it's respective police department. Both for general population and the police department, we classify the representatives in five racial/ethnic categories- White, Black, Hispanic, Asian and Others (which includes insignificant minority representation of Native Americans). Then,
$$D = -\sum_i p_i\log(p_i),$$
where, $i$ denotes the racial/ethnic groups and $p_i$ corresponds to the respective fraction the group occupies in the total population (i.e. given $N$ as the total population size and $N_i$ as the total size of the ethnic group in the population, $p_i = N_i/N$). The function $D$ is a good estimator of diversity as it is,
1. $\textit{scale invariant}$, i.e. independent of proportionate growth of all minority populations simultaneously.
2. Symmetric under the permutations of ethnic groups. (i.e. $D$ is invariant under $P_i\rightarrow P_j$ and vice versa for all $i$ and $j$.)
3. $D' > D$ if $N_i ' = N_i +1$ and $N_j ' = N_j -1$ for distinct $i$ and $j$ provided $p_j > p_i$. 

Throughout our project we have taken $\log$ to mean natural logarithm and hence $D$ is bounded between $0$ and $\log(5)$ (because there are 5 ethnic groups). We let $D_{pol}$ and $D_{pop}$ denote the information entropy of the police department and the population under its jurisdiction, respectively. Since we are interested in understanding the effect of police diversity with repect to the diversity in its jurisdiction. We compute the relative diversity using a slight variation of well known Jensen–Shannon divergence,
$$ R =  sign(D_{pol}-D_{pop})(0.5)\sum_i p_i\log (p_i\frac{p_i+q_i}{2} + p_i\log (p_i\frac{p_i+q_i}{2}).$$
This definition allows us to take advantage of well behaved mathematical properties of Jensen–Shannon divergence like boundedness, positivity and symmetry to make quantitative statements about relative diversity of police department with repect to its jurisdiction, and at the same time allows us to distinguish whether the police department is over or under-represented with repect to the population of its jurisdiction. Over-represented departments have positive value of $R$ whereas, under-represented departments have a negative value of $R$.

## Analysis

```python
import pandas as pd
import numpy as np
import math
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
cleanData = pd.DataFrame.from_csv('clean1R.csv')
cleanData[:7]
```

```python
pop_log = np.log(np.array(cleanData['Population'].tolist()))      #log of population
crm_log = np.log(np.array(cleanData['ViolentCrime'].tolist()))    #log of violent crime
viol_crm = np.array(cleanData['ViolentCrime'].tolist())
pop = np.array(cleanData['Population'].tolist())
prct_crm_percap = 100*viol_crm.astype(float)/pop           #percentage crime rate per capita
cleanData_log = cleanData
cleanData_log['log(Population)'] = pop_log
cleanData_log['log(ViolentCrime)'] = crm_log
cleanData_log['prct_crm_percap'] = crm_log
cleanData_log[['log(Population)','log(ViolentCrime)','PoliceViolence']].boxplot()
```

The data suggests that deadly use of police force are extremely rare events. In fact use of deadly force by police is exponentially small in both population and violent-crime instances. 

```python
cleanData_log[['PoliceViolence']].hist(bins=21)
```

It is clear that instances of police violence can be modelled as an exponential distribution. However, modelling police violence is not the goal, rather we seek to identify if we can successfully identify predictors for the same.
To gain intuition about the expanse of crime in the country and associated police violence we present the following geospatial plot.

```python
def color_gradient(x):
    return (250*(1-x),250*(1-x),255)

color = []    
for i in np.linspace(0,1,100):
    color.append(color_gradient(i))

def color_hex(x):
    return '#%02x%02x%02x' % x
#-------------------------------------------------------------------------------------------------------
range_crime = cleanData_log['prct_crm_percap'].max() - cleanData_log['prct_crm_percap'].min()
min_crime = cleanData_log['prct_crm_percap'].min()
crime_color = []
for i in cleanData_log['prct_crm_percap']:
    crime_color.append(color_hex(color_gradient((i-min_crime)/range_crime)))
    
#-------------------------------------------------------------------------------------------------------
import matplotlib.pyplot as plt
import matplotlib as mpl
%matplotlib inline
# Make a figure and axes with dimensions as desired.
fig = plt.figure(figsize=(8, 3))
ax1 = fig.add_axes([0.05, 0.80, 0.9, 0.15])
# Set the colormap and norm to correspond to the data for which
# the colorbar will be used.
cmap = mpl.cm.Blues
norm = mpl.colors.Normalize(vmin=cleanData_log['prct_crm_percap'].min(), vmax=cleanData_log['prct_crm_percap'].max())

# ColorbarBase derives from ScalarMappable and puts a colorbar
# in a specified axes, so it has everything needed for a
# standalone colorbar.  There are many more kwargs, but the
# following gives a basic continuous colorbar with ticks
# and labels.
cb1 = mpl.colorbar.ColorbarBase(ax1, cmap=cmap,
                                norm=norm,
                                orientation='horizontal')
cb1.set_label('Percentage violent crime per capita')
plt.show()

import folium ; import ast
map_us = folium.Map(location=[29.3, -97],zoom_start=4)
cleanData_log = cleanData_log.dropna()
for i in range(len(cleanData_log)):
    try:
        folium.CircleMarker(location = ast.literal_eval(cleanData_log['Location'][i]), 
                            radius= cleanData_log['Population'][i]/10,
                    popup=cleanData_log['City'][i], color=crime_color[i],
                    fill_color=crime_color[i]).add_to(map_us)
        
    except: 
        print i; raise
map_us
```

We have plotted as plotted percentage violent crime per capita on the map. The percentage violent crime per capita is given by the color bar abpve and the size of the circular marker is proportional to the population size. We observe that high crime areas tend to have higher violent crime percentage per capita. Also, observe that crime prone regions exist in proximity to relatively peaceful regions (depicted by lighter colors). Next we plot instances of police use of deadly force.

```python
import folium ; import ast
map_us = folium.Map(location=[29.3, -97],zoom_start=4)
cleanData_log = cleanData_log.dropna()
for i in range(len(cleanData_log)):
    try:
        folium.CircleMarker(location = ast.literal_eval(cleanData_log['Location'][i]), 
                            radius= cleanData_log['PoliceViolence'][i]*8000,
                    popup=cleanData_log['City'][i], color='#F58282',
                    fill_color='#F58282').add_to(map_us)
        
    except: 
        print i; raise
map_us
```

Police use of deadly force is plotted on the map. We have not used any color gradient to differentiate between rw numbers but instead have scaled the radius with the instances of police shootings. Since we scaled the percentage violent crime per capita with the population size, it appears there is an increasing functional dependence of number of police shootings with the population size. More analysis is needed to find what the functional dependence is and how it relates to crime rates per capita.

Observe that in general populations are more diverse than the respective police departments, as most population diversity lay above the line with slope 1. Next we have histogram relative diversity vs scatter plot of police violence. This plots and the pearsons correlation suggests that,
1. Though most police departments are relatively less diverse than the repective jurisdiction, very few police deparments are heavily skewed.

3. Relative diversity seems to have no correlation with police violence.

```python
#SVR regression
from sklearn.svm import SVR

X = np.array(cleanData_log['prct_crm_percap'].tolist())
Z = np.array(cleanData_log['PoliceViolence'].tolist())
Y = np.array(cleanData_log['R'].tolist())

X_train = np.column_stack((X,Y))
Z_train = Z

svr_rbf = SVR(kernel='rbf', C=1e3, gamma=0.1)
svr_lin = SVR(kernel='linear', C=1e3)
svr_poly = SVR(kernel='poly', C=1e3, degree=2)
Z_rbf = svr_rbf.fit(X_train, Z_train).predict(X_train)
Z_lin = svr_lin.fit(X_train, Z_train).predict(X_train)
Z_poly = svr_poly.fit(X_train, Z_train).predict(X_train)

lw = 2
plt.scatter(X, Z_train, color='darkorange', label='data')
plt.hold('on')
plt.plot(X, Z_rbf, color='navy', lw=lw, label='RBF model')
plt.plot(X, Z_lin, color='c', lw=lw, label='Linear model')
plt.plot(X, Z_poly, color='cornflowerblue', lw=lw, label='Polynomial model')
plt.xlabel('data')
plt.ylabel('target')
plt.title('Support Vector Regression')
plt.legend()
plt.show()
```
